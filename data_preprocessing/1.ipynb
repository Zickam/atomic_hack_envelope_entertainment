```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel(r"../data/dataset/Вопросы и ответы.xlsx", header=1)
```

```python
data = data.drop(columns=["Код", "Время регистрации", "Рабочая группа"])
```

```python
data["Категория"] = data["Категория"].str.lower()
data.sample(10)
```

```python
uniques = list(set(data["Категория"].values))
data["Категория"] = data["Категория"].apply(lambda x: uniques.index(x))
data.head(50)
```

```python
for i in range(1, 4):
    data[f"Аналитика {i}"] = data[f"Аналитика {i}"].fillna("-")
data.head(50)
```

```python
data = data.loc[data["Решение"] != "."]
data
```

```python
data = data.dropna()
data
```

```python
for col in ("Описание", "Решение"):
    item = data[col].apply(len)
    first_filter = item.quantile(0.25)
    second_filter = item.min()

    third_filter = item.quantile(0.75)
    fourth_filter = item.max()

    a, b = (first_filter + second_filter) // 2, (third_filter + fourth_filter) // 2

    data = data.loc[(item > a) & (item < b)]
```

```python
data.to_csv("../data/dataset/after_preprocessing.csv", encoding='utf-8')
```
